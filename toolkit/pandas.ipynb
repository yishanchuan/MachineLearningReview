### 时间序列

```python
n_items = 366
ts = pd.Series(np.random.randn(n_items), index = pd.date_range('20000101', periods=n_items))
print(ts.shape)
ts.head(5)
```

可以将数据按月份聚合

```python
ts.resample("1m").sum()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(10, 6), dpi=144)
cs = ts.cumsum()
cs.plot()
```
